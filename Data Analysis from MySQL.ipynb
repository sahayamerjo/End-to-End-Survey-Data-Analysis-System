```python
import mysql.connector
import pandas as pd
```

```python
# Database Connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="online_learning_experience_survey")
```

```python
# Read data from MySQL
query = "SELECT * FROM response"
Survey = pd.read_sql(query, conn)

conn.close()
```

1.	Total number of responses

```python
Total_responses = len(Survey)
Total_responses
```

2.	Frequency count of each poll option

```python
Survey['poll_option'].value_counts()
```

3.	Percentage distribution of responses

```python
percentage_distribution = (Survey['poll_option'].value_counts(normalize=True) * 100)
percentage_distribution
```

Most and least selected options

```python
most_selected = Survey['poll_option'].value_counts().idxmax()  #it returns the name of the option
most_selected
```

```python
least_selected = Survey['poll_option'].value_counts().idxmin()
least_selected
```

5.	Trend analysis (if timestamps are used)

```python
Survey['timestamp'] = pd.to_datetime(Survey['timestamp'])
Survey.groupby(Survey['timestamp'].dt.date).size()
```

6.	Group-based analysis (e.g., age group vs choice)

```python
group_based = pd.crosstab(Survey['age_group'], Survey['poll_option']) #Compares age group with option
group_based
```
